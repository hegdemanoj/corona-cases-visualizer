```python
districtstate = pd.concat([df['detected_district']+','+df['detected_state']]).unique()
```

```python
print(districtstate)
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="CoronaIndia")
location = geolocator.geocode("Thrissur,Kerala")
print((location.latitude, location.longitude))
```

```python
def findLatLong(locationstr):
    try:
        geolocator = Nominatim(user_agent="CoronaIndia")
        location = geolocator.geocode(locationstr)
        return (location)
    except:
        return None
```

```python
locationMap=dict()
for each in districtstate:
    print("Started:",each)
    loc = findLatLong(each)
    if loc is not None:
        locationMap[each]=[loc.latitude,loc.longitude]
locationMap
```

```python
i=0
while(i<len(df)):
    district = df.iloc[i]['detected_district']
    state = df.iloc[i]['detected_state']
    
    if(type(district)==type(state)==str):
    #     location=findLatLong(district,state)
        loc = district+','+state
        print(i,loc)
        if loc in locationMap.keys():
            df.loc[i,'latitude'] = locationMap[loc][0]
            df.loc[i,'longitude'] = locationMap[loc][1]
    
    i+=1
```
